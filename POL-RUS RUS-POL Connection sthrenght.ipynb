```python
import pandas as pd
import seaborn as sns
sns.set_style("whitegrid")
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style("darkgrid")
```

```python
RUSmonthyearAc2 = pd.read_csv('queries_results/RUS/RUS monthyear ac2 count.csv', header = 0,sep=' *, *')
RUSmonthyearAc2["Time"]=pd.to_datetime(RUSmonthyearAc2['MonthYear'], format='%Y%m').dt.strftime('%Y-%m')

RUSPOLmonthyearERC = pd.read_csv('queries_results/RUS/RUS_POL/RUS-POL monthyear eventRC count SUMaTone AVGaTonne.csv', header = 0,sep=' *, *', dtype={'EventRootCode':str})
RUSPOLmonthyearERC["Time"]=pd.to_datetime(RUSPOLmonthyearERC['MonthYear'], format='%Y%m').dt.strftime('%Y-%m')
eventcodes = pd.read_csv('cameo.csv', header = 0,dtype =str, sep=' *, *')
RUSRUSmonthyearERC = RUSPOLmonthyearERC.merge(eventcodes.set_index('CAMEOEVENTCODE'), left_on='EventRootCode',right_on='CAMEOEVENTCODE', how='left')

s = RUSRUSmonthyearERC.groupby(["Time"]).agg({'Count':'sum'})
t = RUSmonthyearAc2.groupby(["Time"]).agg({'Count':'sum'})
s['Count']=s['Count']/t['Count']
s['Connection']='RUS-POL'
```

```python
POLmonthyearAc2 = pd.read_csv('queries_results/Poland/POL monthyear ac2 count.csv', header = 0,sep=' *, *')
POLmonthyearAc2["Time"]=pd.to_datetime(POLmonthyearAc2['MonthYear'], format='%Y%m').dt.strftime('%Y-%m')

POLmonthyearERC = pd.read_csv('queries_results/Poland/POL_RUS/POL-RUS monthyear eventRC count SUMaTone AVGaTonne.csv', header = 0,sep=' *, *', dtype={'EventRootCode':str})
POLmonthyearERC["Time"]=pd.to_datetime(POLmonthyearERC['MonthYear'], format='%Y%m').dt.strftime('%Y-%m')
eventcodes = pd.read_csv('cameo.csv', header = 0,dtype =str, sep=' *, *')
POLmonthyearERC = POLmonthyearERC.merge(eventcodes.set_index('CAMEOEVENTCODE'), left_on='EventRootCode',right_on='CAMEOEVENTCODE', how='left')
u = POLmonthyearERC.groupby(["Time"]).agg({'Count':'sum'})
t = POLmonthyearAc2.groupby(["Time"]).agg({'Count':'sum'})

u['Count']=u['Count']/t['Count']
u['Connection']='POL-RUS'

s = s.append(u)
```

```python
s = s.groupby(["Time","Connection"]).agg({'Count':'sum'})
g = s.unstack().plot(y='Count')
g
g.set(ylabel='Count')
g.set_title("Connection strength since 2015")
g.figure.set_size_inches(20,8)
```
