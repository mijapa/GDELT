```python
import pandas as pd
import seaborn as sns
sns.set_style("whitegrid")
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style("darkgrid")
GLyearAc1 = pd.read_csv('queries_results/global/GLOBAL year ac1 count.csv', header = 0,sep=' *, *')
```

```python
s = GLyearAc1.groupby("Actor1CountryCode").sum().sort_values(by="Count", ascending=False).head(20)
g = sns.barplot(s.index, s.Count, palette='rainbow')
g.set_title("Actor 1 Country Code since 2015")
g.set(ylabel='Count')
g.figure.set_size_inches(20,8)
```

```python
s = GLyearAc1.groupby("Actor1CountryCode").sum().sort_values(by="Count", ascending=False).head(20)[1:]
g = sns.barplot(s.index, s.Count, palette='rainbow')
g.set_title("Actor 1 Country Code since 2015 (after USA)")
g.set(ylabel='Count')
g.figure.set_size_inches(20,8)
```

```python
fig, ax =plt.subplots(1,5)

s = GLyearAc1[GLyearAc1.Year.eq(2015)].groupby("Actor1CountryCode").sum().sort_values(by="Count", ascending=False).head(8)
g = sns.barplot(s.index, s.Count, palette='rainbow', ax=ax[0])
g.set_title("2015")

s = GLyearAc1[GLyearAc1.Year.eq(2016)].groupby("Actor1CountryCode").sum().sort_values(by="Count", ascending=False).head(8)
g = sns.barplot(s.index, s.Count, palette='rainbow', ax=ax[1])
g.set_title("2016")

s = GLyearAc1[GLyearAc1.Year.eq(2017)].groupby("Actor1CountryCode").sum().sort_values(by="Count", ascending=False).head(8)
g = sns.barplot(s.index, s.Count, palette='rainbow', ax=ax[2])
g.set_title("2017")

s = GLyearAc1[GLyearAc1.Year.eq(2018)].groupby("Actor1CountryCode").sum().sort_values(by="Count", ascending=False).head(8)
g = sns.barplot(s.index, s.Count, palette='rainbow', ax=ax[3])
g.set_title("2018")

s = GLyearAc1[GLyearAc1.Year.eq(2019)].groupby("Actor1CountryCode").sum().sort_values(by="Count", ascending=False).head(8)
g = sns.barplot(s.index, s.Count, palette='rainbow', ax=ax[4])
g.set_title("2019")

fig.suptitle("Popularity of Actor 1 Country over years")
g.figure.set_size_inches(20,8)
```

```python
fig, ax =plt.subplots(1,5)

s = GLyearAc1[GLyearAc1.Year.eq(2015)].groupby("Actor1CountryCode").sum().sort_values(by="Count", ascending=False).head(8)[1:]
g = sns.barplot(s.index, s.Count, palette='rainbow', ax=ax[0])
g.set_title("2015")

s = GLyearAc1[GLyearAc1.Year.eq(2016)].groupby("Actor1CountryCode").sum().sort_values(by="Count", ascending=False).head(8)[1:]
g = sns.barplot(s.index, s.Count, palette='rainbow', ax=ax[1])
g.set_title("2016")

s = GLyearAc1[GLyearAc1.Year.eq(2017)].groupby("Actor1CountryCode").sum().sort_values(by="Count", ascending=False).head(8)[1:]
g = sns.barplot(s.index, s.Count, palette='rainbow', ax=ax[2])
g.set_title("2017")

s = GLyearAc1[GLyearAc1.Year.eq(2018)].groupby("Actor1CountryCode").sum().sort_values(by="Count", ascending=False).head(8)[1:]
g = sns.barplot(s.index, s.Count, palette='rainbow', ax=ax[3])
g.set_title("2018")

s = GLyearAc1[GLyearAc1.Year.eq(2019)].groupby("Actor1CountryCode").sum().sort_values(by="Count", ascending=False).head(8)[1:]
g = sns.barplot(s.index, s.Count, palette='rainbow', ax=ax[4])
g.set_title("2019")

fig.suptitle("Popularity of Actor 1 Country over years (after USA)")
g.figure.set_size_inches(20,8)
```

```python
n=10
cols = GLyearAc1.groupby("Actor1CountryCode", as_index=False).sum().sort_values(by="Count", ascending=False).head(n)["Actor1CountryCode"]
s  = GLyearAc1.loc[GLyearAc1["Actor1CountryCode"].isin(cols)]

s = s.groupby(["Year", "Actor1CountryCode"]).agg({'Count':'sum'})
g = s.unstack().plot(y='Count')
g.set_title("Popularity of Actor 1 Country since 2015 - top {}".format(n))
g.set(ylabel='Count')
g.figure.set_size_inches(20,8)
plt.savefig('text/fig/GLOBAL/Actor1inTIME.png', bbox_inches='tight')
```

```python
cols = GLyearAc1.groupby("Actor1CountryCode", as_index=False).sum().sort_values(by="Count", ascending=False).head(n)[1:]["Actor1CountryCode"]
s  = GLyearAc1.loc[GLyearAc1["Actor1CountryCode"].isin(cols)]

s = s.groupby(["Year", "Actor1CountryCode"]).agg({'Count':'sum'}).sort_values(by=["Year","Count"], ascending=False)
g = s.unstack().plot(y='Count')
g.set(ylabel='Count')
g.set_title("Popularity of Actor 1 Country since 2015 (after USA) - top {}".format(n))
g.figure.set_size_inches(20,8)
plt.savefig('text/fig/GLOBAL/Actor1inTIMEafterUSA.png', bbox_inches='tight')
```
