```python
import pandas as pd
import seaborn as sns
sns.set_style("whitegrid")
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style("darkgrid")
```

```python
USAmonthyearAc2 = pd.read_csv('queries_results/USA/USA monthyear ac2 count.csv', header = 0,sep=' *, *')
USAmonthyearAc2["Time"]=pd.to_datetime(USAmonthyearAc2['MonthYear'], format='%Y%m').dt.strftime('%Y-%m')

USAPOLmonthyearERC = pd.read_csv('queries_results/USA/USA_POL/USA-POL monthyear eventRC count SUMaTone AVGaTonne.csv', header = 0,sep=' *, *', dtype={'EventRootCode':str})
USAPOLmonthyearERC["Time"]=pd.to_datetime(USAPOLmonthyearERC['MonthYear'], format='%Y%m').dt.strftime('%Y-%m')
eventcodes = pd.read_csv('cameo.csv', header = 0,dtype =str, sep=' *, *')
USAUSAmonthyearERC = USAPOLmonthyearERC.merge(eventcodes.set_index('CAMEOEVENTCODE'), left_on='EventRootCode',right_on='CAMEOEVENTCODE', how='left')

s = USAUSAmonthyearERC.groupby(["Time"]).agg({'Count':'sum'})
t = USAmonthyearAc2.groupby(["Time"]).agg({'Count':'sum'})

s['Count']=s['Count']/t['Count']
s['Connection']='USA-POL'
```

```python
POLmonthyearAc2 = pd.read_csv('queries_results/Poland/POL monthyear ac2 count.csv', header = 0,sep=' *, *')
POLmonthyearAc2["Time"]=pd.to_datetime(POLmonthyearAc2['MonthYear'], format='%Y%m').dt.strftime('%Y-%m')

POLUSAmonthyearERC = pd.read_csv('queries_results/Poland/POL_USA/POL-USA monthyear eventRC count SUMaTone AVGaTonne.csv', header = 0,sep=' *, *', dtype={'EventRootCode':str})
POLUSAmonthyearERC["Time"]=pd.to_datetime(POLUSAmonthyearERC['MonthYear'], format='%Y%m').dt.strftime('%Y-%m')
eventcodes = pd.read_csv('cameo.csv', header = 0,dtype =str, sep=' *, *')
POLUSAmonthyearERC = POLUSAmonthyearERC.merge(eventcodes.set_index('CAMEOEVENTCODE'), left_on='EventRootCode',right_on='CAMEOEVENTCODE', how='left')

u = POLUSAmonthyearERC.groupby(["Time"]).agg({'Count':'sum'})
t = POLmonthyearAc2.groupby(["Time"]).agg({'Count':'sum'})

u['Count']=u['Count']/t['Count']
u['Connection']='POL-USA'

s = s.append(u)
```

```python
s = s.groupby(["Time","Connection"]).agg({'Count':'sum'})
g = s.unstack().plot(y='Count')
g
g.set(ylabel='Count')
g.set_title("Connection strength since 2015")
g.figure.set_size_inches(20,8)
```
