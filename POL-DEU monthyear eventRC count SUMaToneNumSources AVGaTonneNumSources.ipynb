```python
import pandas as pd
import seaborn as sns
sns.set_style("whitegrid")
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style("darkgrid")
POLmonthyearERC = pd.read_csv('queries_results/Poland/POL_DEU/POL-DEU monthyear eventRC count SUMaToneNumSources AVGaTonneNumSources.csv', header = 0,sep=' *, *', dtype={'EventRootCode':str})
POLmonthyearERC["Time"]=pd.to_datetime(POLmonthyearERC['MonthYear'], format='%Y%m').dt.strftime('%Y-%m')
eventcodes = pd.read_csv('cameo.csv', header = 0,dtype =str, sep=' *, *')
POLmonthyearERC = POLmonthyearERC.merge(eventcodes.set_index('CAMEOEVENTCODE'), left_on='EventRootCode',right_on='CAMEOEVENTCODE', how='left')
```

```python
s = POLmonthyearERC.groupby(["Time","EVENTDESCRIPTION"]).agg({'SumAvgToneNumSources':'sum'})

g = s.unstack().plot(y='SumAvgToneNumSources')

g.set(ylabel='Summary AvgTone')
g.set_title("Summary AvgTone times NumSources for POL-DEU since 2015")
g.figure.set_size_inches(20,8)
```

```python
s = POLmonthyearERC.groupby(["Time","EVENTDESCRIPTION"]).agg({'AvgAvgToneNumSources':'sum'})
g = s.unstack().plot(y='AvgAvgToneNumSources')
g.set(ylabel='Average AvgTone')
g.set_title("Average AvgTone times NumSources for POL-DEU since 2015")
g.figure.set_size_inches(20,8)
```
