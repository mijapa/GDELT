```python
import pandas as pd
import seaborn as sns
sns.set_style("whitegrid")
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style("darkgrid")
DEUmonthyearERC = pd.read_csv('queries_results/DEU/DEU monthyear eventRC count.csv', header = 0,sep=' *, *', dtype={'EventRootCode':str})
DEUmonthyearERC["Time"]=pd.to_datetime(DEUmonthyearERC['MonthYear'], format='%Y%m').dt.strftime('%Y-%m')
eventcodes = pd.read_csv('cameo.csv', header = 0,dtype =str, sep=' *, *')
DEUmonthyearERC = DEUmonthyearERC.merge(eventcodes.set_index('CAMEOEVENTCODE'), left_on='EventRootCode',right_on='CAMEOEVENTCODE', how='left')
```

```python
s = DEUmonthyearERC.groupby("EVENTDESCRIPTION", as_index=False).sum().sort_values(by="Count", ascending=False).head(30)
g = sns.barplot(s.EVENTDESCRIPTION, s.Count, palette='rainbow')
g.set_title("Event Root Code count for DEU since 2015")
g.set(ylabel='Count')
g.set_xticklabels(g.get_xticklabels(), rotation=40, ha="right")
g.figure.set_size_inches(20,8)
plt.savefig('text/fig/DEU/DEUERC.png', bbox_inches='tight')
```

```python
ile = 10
cols = DEUmonthyearERC.groupby("EVENTDESCRIPTION", as_index=False).sum().sort_values(by="Count", ascending=False).head(ile)['EVENTDESCRIPTION']
s  = DEUmonthyearERC.loc[DEUmonthyearERC["EVENTDESCRIPTION"].isin(cols)]

s = s.groupby(["Time","EVENTDESCRIPTION"]).agg({'Count':'sum'})
g = s.unstack().plot(y='Count')
g.set(ylabel='Count')
g.set_title("Event Root Code count for DEU since 2015 - top {}".format(ile))
g.figure.set_size_inches(20,8)
plt.savefig('text/fig/DEU/DEUERCinTIME.png', bbox_inches='tight')
```

```python
s['%']= s.groupby(level=0).apply(lambda x:  100*x / x.sum())

g = s.unstack().plot(y='%')
g.set(ylabel='Count %')
g.set_title("Event Root Code count for DEU since 2015 - top {}".format(ile))
g.figure.set_size_inches(20,8)
plt.savefig('text/fig/DEU/DEUERCpercinTIME.png', bbox_inches='tight')
```
