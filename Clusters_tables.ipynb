```python
merged['gdp_per_cap'] = merged.gdp_md_est / merged.pop_est

from IPython.display import display

def df_to_png(df, file):
    import imgkit
    imgkit.from_string(df.render(), file,
                       options={
            'format': 'png',
            'quality' : '100',
            'width' : '800',
            'quiet' : ''
                       })

for c in range(10):
    display('cluster {}'.format(c))
    
    cluster_tab = merged[merged.cluster==c][['iso_a3', 'name', 'continent' ,'pop_est', 'gdp_per_cap']].sort_values(by='name')
    cluster_tab['pop_est']=cluster_tab['pop_est'].map('{:,}'.format)
    cluster_tab=cluster_tab\
        .style\
        .hide_index()
    display(cluster_tab)

    df_to_png(cluster_tab, 'text/tables/CLUST/clust{}kmeans.png'.format(c))
```
