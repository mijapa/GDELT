```python
import pandas as pd
import seaborn as sns
sns.set_style("whitegrid")
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style("darkgrid")
POLmonthyearAc1 = pd.read_csv('queries_results/Poland/POL monthyear ac2 fitht count.csv', header = 0, sep=' *, *')
POLmonthyearAc1["Time"]=pd.to_datetime(POLmonthyearAc1['MonthYear'], format='%Y%m').dt.strftime('%Y-%m')
```

```python
g = sns.lineplot(x="Time", y="FightCountries", marker="o", dashes=True, data=POLmonthyearAc1)
g.set_title("Fight Countries Count for Poland")
g.set(ylabel='Count')
g.figure.set_size_inches(20,8)
```
