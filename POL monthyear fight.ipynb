```python
import pandas as pd
import seaborn as sns
sns.set_style("whitegrid")
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style("darkgrid")
GLOBALmonthyearAc1 = pd.read_csv('queries_results/global/GLOBAL monthyear ac2 fitht count.csv', header = 0, sep=' *, *')
GLOBALmonthyearAc1["Time"]=pd.to_datetime(GLOBALmonthyearAc1['MonthYear'], format='%Y%m').dt.strftime('%Y-%m')
```

```python
POLmonthyearAc1 = pd.read_csv('queries_results/Poland/POL monthyear ac2 fitht count.csv', header = 0, sep=' *, *')
POLmonthyearAc1["Time"]=pd.to_datetime(POLmonthyearAc1['MonthYear'], format='%Y%m').dt.strftime('%Y-%m')
POLmonthyearAc1 = POLmonthyearAc1.\
    merge(GLOBALmonthyearAc1, on="Time", suffixes=('','Global'))
POLmonthyearAc1['FightCountriesPercentage']=POLmonthyearAc1['FightCountries']/POLmonthyearAc1['FightCountriesGlobal']*100
```

```python
s = POLmonthyearAc1
s = s.groupby(["Time"]).agg({'FightCountries':'sum'})
g = s.plot()
g.set_xticklabels(g.get_xticklabels(), rotation=40, ha="right")
g.set_title("ERC Fight countries count for POL since 2015")
g.set(ylabel='Count')
g.figure.set_size_inches(20,8)
```

```python
s = POLmonthyearAc1
s = s.groupby(["Time"]).agg({'FightCountriesPercentage':'sum'})
g = s.plot()
g.set_title("ERC Fight countries count for POL since 2015")
g.set(ylabel='Percentage')
g.set(xlabel='Time')
g.set_xticklabels(g.get_xticklabels(), rotation=40, ha="right")
g.figure.set_size_inches(20,8)
```
