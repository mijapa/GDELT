```python
import pandas as pd
import seaborn as sns
sns.set_style("whitegrid")
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
%matplotlib inline

sns.set_style("darkgrid")
USAmonthyearAc2 = pd.read_csv('queries_results/USA/USA monthyear ac2 count.csv', header = 0,sep=' *, *')
USAmonthyearAc2["Time"]=pd.to_datetime(USAmonthyearAc2['MonthYear'], format='%Y%m').dt.strftime('%Y-%m')
```

```python
s = USAmonthyearAc2.groupby("Actor2CountryCode").sum().sort_values(by="Count", ascending=False).head(20)
g = sns.barplot(s.index, s.Count, palette='rainbow')
g.set_title("Actor2 Country Code pair for USA as Actor1 since 2015")
g.set(ylabel='Count')
g.figure.set_size_inches(20,8)
plt.savefig('text/fig/USA/USAactor2Pair.png', bbox_inches='tight')
```

```python
s = USAmonthyearAc2.groupby("Actor2CountryCode").sum().sort_values(by="Count", ascending=False).head(20)[1:]
g = sns.barplot(s.index, s.Count, palette='rainbow')
g.set_title("Actor2 Country Code pair for USA as Actor1 since 2015 (after USA)")
g.set(ylabel='Count')
g.figure.set_size_inches(20,8)
```

```python
ile = 8
cols = USAmonthyearAc2.groupby("Actor2CountryCode", as_index=False).sum().sort_values(by="Count", ascending=False).head(ile)['Actor2CountryCode']
s  = USAmonthyearAc2.loc[USAmonthyearAc2["Actor2CountryCode"].isin(cols)]
s = s.groupby(["Time","Actor2CountryCode"]).agg({'Count':'sum'})

g = s.unstack().plot(y='Count')
g.set(ylabel='Count')
g.set_title("Popularity of Actor2 Country pair for USA as Actor1 since 2015 - top {}".format(ile))
g.figure.set_size_inches(20,8)
```

```python
s['%']= s.groupby(level=0).apply(lambda x:  100*x / x.sum())

g = s.unstack().plot(y='%')
g.set(ylabel='Count %')
g.set_title("Popularity of Actor2 Country pair for USA as Actor1 since 2015 - top {}".format(ile))
g.figure.set_size_inches(20,8)
plt.savefig('text/fig/USA/USAactor2PairPercinTIME.png', bbox_inches='tight')
```
