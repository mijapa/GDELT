```python
import pandas as pd
import seaborn as sns
sns.set_style("whitegrid")
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style("darkgrid")
GLOBALmonthyearAc1 = pd.read_csv('queries_results/global/GLOBAL monthyear ac2 fitht count.csv', header = 0, sep=' *, *')
GLOBALmonthyearAc1["Time"]=pd.to_datetime(GLOBALmonthyearAc1['MonthYear'], format='%Y%m').dt.strftime('%Y-%m')
```

```python
s = GLOBALmonthyearAc1
s = s.groupby(["Time"]).agg({'FightCountries':'sum'})
g = s.plot()
g.set_xticklabels(g.get_xticklabels(), rotation=40, ha="right")
g.set_title("ERC Fight countries count since 2015")
g.set(ylabel='Count')
g.figure.set_size_inches(20,8)
```

```python
CHNmonthyearAc1 = pd.read_csv('queries_results/CHN/CHN monthyear ac2 fitht count.csv', header = 0, sep=' *, *')
CHNmonthyearAc1["Time"]=pd.to_datetime(CHNmonthyearAc1['MonthYear'], format='%Y%m').dt.strftime('%Y-%m')
CHNmonthyearAc1 = CHNmonthyearAc1.\
    merge(GLOBALmonthyearAc1, on="Time", suffixes=('','Global'))
CHNmonthyearAc1['FightCountriesPercentage']=CHNmonthyearAc1['FightCountries']/CHNmonthyearAc1['FightCountriesGlobal']*100
CHNmonthyearAc1['Country']="CHN"
```

```python
DEUmonthyearAc1 = pd.read_csv('queries_results/DEU/DEU monthyear ac2 fitht count.csv', header = 0, sep=' *, *')
DEUmonthyearAc1["Time"]=pd.to_datetime(DEUmonthyearAc1['MonthYear'], format='%Y%m').dt.strftime('%Y-%m')
DEUmonthyearAc1 = DEUmonthyearAc1.\
    merge(GLOBALmonthyearAc1, on="Time", suffixes=('','Global'))
DEUmonthyearAc1['FightCountriesPercentage']=DEUmonthyearAc1['FightCountries']/DEUmonthyearAc1['FightCountriesGlobal']*100
DEUmonthyearAc1['Country']="DEU"
```

```python
POLmonthyearAc1 = pd.read_csv('queries_results/Poland/POL monthyear ac2 fitht count.csv', header = 0, sep=' *, *')
POLmonthyearAc1["Time"]=pd.to_datetime(POLmonthyearAc1['MonthYear'], format='%Y%m').dt.strftime('%Y-%m')
POLmonthyearAc1 = POLmonthyearAc1.\
    merge(GLOBALmonthyearAc1, on="Time", suffixes=('','Global'))
POLmonthyearAc1['FightCountriesPercentage']=POLmonthyearAc1['FightCountries']/POLmonthyearAc1['FightCountriesGlobal']*100
POLmonthyearAc1['Country']="POL"
```

```python
RUSmonthyearAc1 = pd.read_csv('queries_results/RUS/RUS monthyear ac2 fitht count.csv', header = 0, sep=' *, *')
RUSmonthyearAc1["Time"]=pd.to_datetime(RUSmonthyearAc1['MonthYear'], format='%Y%m').dt.strftime('%Y-%m')
RUSmonthyearAc1 = RUSmonthyearAc1.\
    merge(GLOBALmonthyearAc1, on="Time", suffixes=('','Global'))
RUSmonthyearAc1['FightCountriesPercentage']=RUSmonthyearAc1['FightCountries']/RUSmonthyearAc1['FightCountriesGlobal']*100
RUSmonthyearAc1['Country']="RUS"
```

```python
SAUmonthyearAc1 = pd.read_csv('queries_results/SAU/SAU monthyear ac2 fitht count.csv', header = 0, sep=' *, *')
SAUmonthyearAc1["Time"]=pd.to_datetime(SAUmonthyearAc1['MonthYear'], format='%Y%m').dt.strftime('%Y-%m')
SAUmonthyearAc1 = SAUmonthyearAc1.\
    merge(GLOBALmonthyearAc1, on="Time", suffixes=('','Global'))
SAUmonthyearAc1['FightCountriesPercentage']=SAUmonthyearAc1['FightCountries']/SAUmonthyearAc1['FightCountriesGlobal']*100
SAUmonthyearAc1['Country']="SAU"
```

```python
USAmonthyearAc1 = pd.read_csv('queries_results/USA/USA monthyear ac2 fitht count.csv', header = 0, sep=' *, *')
USAmonthyearAc1["Time"]=pd.to_datetime(USAmonthyearAc1['MonthYear'], format='%Y%m').dt.strftime('%Y-%m')
USAmonthyearAc1 = USAmonthyearAc1.\
    merge(GLOBALmonthyearAc1, on="Time", suffixes=('','Global'))
USAmonthyearAc1['FightCountriesPercentage']=USAmonthyearAc1['FightCountries']/USAmonthyearAc1['FightCountriesGlobal']*100
USAmonthyearAc1['Country']="USA"
```

```python
gl = DEUmonthyearAc1\
    .append(CHNmonthyearAc1)\
    .append(POLmonthyearAc1)\
    .append(RUSmonthyearAc1)\
    .append(SAUmonthyearAc1)\
    .append(USAmonthyearAc1)
```

```python
s = gl
s = s.groupby(["Time","Country"]).agg({'FightCountries':'sum'})
g = s.unstack().plot()
g.set_xticklabels(g.get_xticklabels(), rotation=40, ha="right")
g.set_title("ERC Fight countries count for DEU since 2015")
g.set(ylabel='Count')
g.figure.set_size_inches(20,8)
```

```python
s = gl
s = s.groupby(["Time", "Country"]).agg({'FightCountriesPercentage':'sum'})
g = s.unstack().plot()
g.set_title("ERC Fight countries count for DEU since 2015")
g.set(ylabel='Percentage')
g.set(xlabel='Time')
g.set_xticklabels(g.get_xticklabels(), rotation=40, ha="right")
g.figure.set_size_inches(20,8)
```
