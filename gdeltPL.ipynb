```python
import pandas as pd
import seaborn as sns
sns.set_style("whitegrid")
%matplotlib inline

sns.set_style("darkgrid")

plAc1 = pd.read_csv('queries_results/count ac1 POL group year.csv', header = 0)
plAc2 = pd.read_csv('queries_results/count ac2 POL group year.csv', header = 0)
plGeo = pd.read_csv('queries_results/count actionGeo POL group year.csv', header = 0)
plAc1.Count.head(1)
```

```python
g = sns.lineplot(x="Year", y="Count", marker="o", dashes=True, data=plAc1)
g.set_title("Popularity of Poland as Actor 1")
g.set(ylabel='Count')
g.figure.set_size_inches(20,8)
```

```python
g = sns.lineplot(x="Year", y="Count", marker="o", dashes=True, data=plAc2)
g.set_title("Popularity of Poland as Actor 2")
g.set(ylabel='Count')
g.figure.set_size_inches(20,8)
```

```python
g = sns.lineplot(x="Year", y="Count", marker="o", dashes=True, data=plGeo)
g.set_title("Popularity of Poland as Location of Events")
g.set(ylabel='Count')
g.figure.set_size_inches(20,8)
```
